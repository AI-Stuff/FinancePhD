```python
%matplotlib inline
import pandas as pd
import seaborn as sns

import pylab
sns.set(font_scale=3)
pylab.rcParams['figure.figsize'] = (24, 18)
```

```python
x=pd.read_csv('FHT.csv',index_col=0)
print x.T.to_latex()
```

```python
y=pd.read_csv('amihud.csv',index_col=0)
print y.T.to_latex()
```

```python
df=pd.DataFrame()
df['Amihud-Mean']=x.Mean
df['Amihud-Std']=x['Standard Deviation']
df['FHT-Mean']=y.Mean
df['FHT-Std']=y['Standard Deviation']
```

```python
df[['FHT-Mean','Amihud-Mean']].plot(kind='bar',secondary_y='FHT-Mean')
```

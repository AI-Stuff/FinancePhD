```python
COT_long=pd.DataFrame()
COT_short=pd.DataFrame()
COT_OI=pd.DataFrame()
for c in CME_contracts:
    ticker='CFTC/'+c+'_F_L_ALL'
    try:
        d=quandl.get(ticker,authtoken=token)
        COT_long[map[c]]= d['Noncommercial Long']
        COT_short[map[c]]= d['Noncommercial Short']
        COT_OI[map[c]]= d['Open Interest']
    except:
        print c
```

```python
COT=pd.DataFrame()
for c in CME_contracts:
    ticker='CFTC/'+c+'_F_L_ALL'
    try:
        d=quandl.get(ticker,authtoken=token)
        x=(d['Noncommercial Long']-d['Noncommercial Short'])
        COT[map[c]]=x/(d['Open Interest'])
    except:
        print c
```

```python
cleansed=load_amihud_markets_price()
amihud=calculate_amihud_liquidity(cleansed)[:'2016'] 
```

```python
ind=amihud.dropna(how='all').index
mon=(COT.resample(rule='m',how='last')/COT.resample(rule='m',how='last').abs()).ix[ind]
```

```python
LG=pd.Series()
SH=pd.Series()
for r in mon.iterrows():
    LG[r[0]]=amihud[r[1][r[1]>0].index].ix[r[0]].dropna().median()
    SH[r[0]]=amihud[r[1][r[1]<0].index].ix[r[0]].dropna().median()
liquidity=pd.DataFrame()
liquidity['Speculators Long']=LG
liquidity['Speculators Short']=SH
```

```python
COT_OI=pd.DataFrame()
for c in CME_contracts:
    ticker='CFTC/'+c+'_F_L_ALL'
    try:
        d=quandl.get(ticker,authtoken=token)
        x=(d['Noncommercial Long']+d['Noncommercial Short'])
        COT_OI[map[c]]=x/(d['Commercial Long']+d['Commercial Short'])
    except:
        print c
```

```python
COT_OI.plot(colormap='jet')
```

```python
COT_OIm=COT_OI.resample(rule='m',how='last')
illiq=pd.Series()
liq=pd.Series()
for i,r in COT_OIm['2000':'2016'].iterrows():
    il=r[r>r.median()].index
    l=r[r<r.median()].index   
    illiq[i]=amihud[il].ix[i].median()
    liq[i]=amihud[l].ix[i].median()
```

# Specualtor ratio

$$ \frac{\sum Spectulator_{Long} + Speculator_{Short} }{\sum Hedger_{Long} + Hedger_{Short}}$$

```python
spec=pd.DataFrame()
spec['Higher Speculator Ratio']=illiq
spec['Lower Speculator Ratio']=liq
spec.plot(colormap='jet')
```
